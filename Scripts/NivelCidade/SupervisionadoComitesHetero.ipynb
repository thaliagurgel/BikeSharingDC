```python
import pandas as pd
# bikesDiscretizado1                   Base3
# bikesDiscretizado2                   Base4
# bikesDiscretizado1Normalizado        Base5
# bikesDiscretizado2Normalizado        Base6
# bikesDiscretizado1Normalizado25      Base7
# bikesDiscretizado1Normalizado50      Base8
# bikesDiscretizado2Normalizado25      Base9
# bikesDiscretizado2Normalizado50      Base10

arquivo = 'Discretizado1Normalizado'
CaminhoArquivo = 'C:/Users/John/Desktop/Mestrado/Dissertacao/ProjetoBikes/Experimentos/Cidade/bikes' + arquivo +'.csv'

base = pd.read_csv(CaminhoArquivo, sep=';');
base.head()
```

```python
import winsound
def FimAlgoritmo():
    winsound.Beep(1500,500);
    winsound.Beep(1500,500);

def FimMetrica():
    winsound.Beep(1000,200);
    
def FimBase():
    winsound.Beep(2000,1000);
    
from sklearn import model_selection
```

```python
# from sklearn import model_selection
from sklearn.model_selection import cross_val_score
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from mlxtend.classifier import StackingClassifier
import warnings
warnings.simplefilter("ignore")

Y = base['qtd']
X = base.drop('qtd', axis=1)
seed = 7

# kfold=model_selection.KFold(n_splits=10, random_state=seed)
clf1 = MLPClassifier(hidden_layer_sizes=(200, ))
clf2 = RandomForestClassifier()
clf3 = DecisionTreeClassifier(random_state=seed)

print("\n\n MLP e RFC:");
print("Meta-classficador: MLP");
model1 = StackingClassifier(classifiers=[clf1,clf2], meta_classifier=clf1)
results = cross_val_score(model1, X, Y, cv=10, scoring='accuracy')
print('Acurácia: %.4f' % results.mean())
print('DesvioP: %.4f' % results.std())
FimMetrica
results = cross_val_score(model1, X, Y, cv=10, scoring='precision_macro')
print('Precisão: %.4f' % results.mean())
FimMetrica()
results = cross_val_score(model1, X, Y, cv=10, scoring='recall_macro')
print('Recall: %.4f \n ' % results.mean())
FimAlgoritmo()

print("Meta-classficador: DT");
model2 = StackingClassifier(classifiers=[clf1,clf2], meta_classifier=clf3)
results = cross_val_score(model2, X, Y, cv=10, scoring='accuracy')
print('Acurácia: %.4f' % results.mean())
print('DesvioP: %.4f' % results.std())
FimMetrica
results = cross_val_score(model2, X, Y, cv=10, scoring='precision_macro')
print('Precisão: %.4f' % results.mean())
FimMetrica()
results = cross_val_score(model2, X, Y, cv=10, scoring='recall_macro')
print('Recall: %.4f \n ' % results.mean())
FimAlgoritmo()


print("\n\n MLP e DT:");
print("Meta-classficador: MLP");
model1 = StackingClassifier(classifiers=[clf1,clf3], meta_classifier=clf1)
results = cross_val_score(model1, X, Y, cv=10, scoring='accuracy')
print('Acurácia: %.4f' % results.mean())
print('DesvioP: %.4f' % results.std())
FimMetrica
results = cross_val_score(model1, X, Y, cv=10, scoring='precision_macro')
print('Precisão: %.4f' % results.mean())
FimMetrica()
results = cross_val_score(model1, X, Y, cv=10, scoring='recall_macro')
print('Recall: %.4f \n ' % results.mean())
FimAlgoritmo()

print("Meta-classficador: DT");
model2 = StackingClassifier(classifiers=[clf1,clf3], meta_classifier=clf3)
results = cross_val_score(model2, X, Y, cv=10, scoring='accuracy')
print('Acurácia: %.4f' % results.mean())
print('DesvioP: %.4f' % results.std())
FimMetrica
results = cross_val_score(model2, X, Y, cv=10, scoring='precision_macro')
print('Precisão: %.4f' % results.mean())
FimMetrica()
results = cross_val_score(model2, X, Y, cv=10, scoring='recall_macro')
print('Recall: %.4f \n ' % results.mean())
FimAlgoritmo()



print("\n\n RFC e DT:");
print("Meta-classficador: MLP");
model1 = StackingClassifier(classifiers=[clf2,clf3], meta_classifier=clf1)
results = cross_val_score(model1, X, Y, cv=10, scoring='accuracy')
print('Acurácia: %.4f' % results.mean())
print('DesvioP: %.4f' % results.std())
FimMetrica
results = cross_val_score(model1, X, Y, cv=10, scoring='precision_macro')
print('Precisão: %.4f' % results.mean())
FimMetrica()
results = cross_val_score(model1, X, Y, cv=10, scoring='recall_macro')
print('Recall: %.4f \n ' % results.mean())
FimAlgoritmo()

print("Meta-classficador: DT");
model2 = StackingClassifier(classifiers=[clf2,clf3], meta_classifier=clf3)
results = cross_val_score(model2, X, Y, cv=10, scoring='accuracy')
print('Acurácia: %.4f' % results.mean())
print('DesvioP: %.4f' % results.std())
FimMetrica
results = cross_val_score(model2, X, Y, cv=10, scoring='precision_macro')
print('Precisão: %.4f' % results.mean())
FimMetrica()
results = cross_val_score(model2, X, Y, cv=10, scoring='recall_macro')
print('Recall: %.4f \n ' % results.mean())
FimAlgoritmo()


print("\n\n MLP, RFC e DT:");
print("Meta-classficador: MLP");
model1 = StackingClassifier(classifiers=[clf1,clf2,clf3], meta_classifier=clf1)
results = cross_val_score(model1, X, Y, cv=10, scoring='accuracy')
print('Acurácia: %.4f' % results.mean())
print('DesvioP: %.4f' % results.std())
FimMetrica
results = cross_val_score(model1, X, Y, cv=10, scoring='precision_macro')
print('Precisão: %.4f' % results.mean())
FimMetrica()
results = cross_val_score(model1, X, Y, cv=10, scoring='recall_macro')
print('Recall: %.4f \n ' % results.mean())
FimAlgoritmo()

print("Meta-classficador: DT");
model2 = StackingClassifier(classifiers=[clf1,clf2,clf3], meta_classifier=clf3)
results = cross_val_score(model2, X, Y, cv=10, scoring='accuracy')
print('Acurácia: %.4f' % results.mean())
print('DesvioP: %.4f' % results.std())
FimMetrica
results = cross_val_score(model2, X, Y, cv=10, scoring='precision_macro')
print('Precisão: %.4f' % results.mean())
FimMetrica()
results = cross_val_score(model2, X, Y, cv=10, scoring='recall_macro')
print('Recall: %.4f \n ' % results.mean())
FimAlgoritmo()

FimBase()
```
