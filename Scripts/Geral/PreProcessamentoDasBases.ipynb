```python
# Função para normalização
# base = nome da base
# var_classe = coluna da classe que não precisa ser normalizada
# numTotal = quantidade todal de registros
def normaliza (base, var_classe, numTotal):
    from sklearn import preprocessing
    min_max_scaler = preprocessing.MinMaxScaler()
    nomeColunas = base.columns.tolist()
    nomeColunas.remove(str(var_classe))
    count=0    
    while (count < len(nomeColunas)):
        var = base[nomeColunas[count]].values
        var = var.reshape((numTotal,1))
        base[nomeColunas[count]] = min_max_scaler.fit_transform(var)
        count+=1
    return base;
```

```python
bikesDiscretizado1Normalizado = pd.read_csv('C:/Users/John/Desktop/Mineracao/projetoBikes/bikesDiscretizado1.csv', sep=';');
# normaliza atributos da base e exporta
bikesDiscretizado1Normalizado = normaliza(bikesDiscretizado1Normalizado, 'qtd', 8737)
bikesDiscretizado1Normalizado.to_csv('bikesDiscretizado1Normalizado.csv', sep=';', index=False)
bikesDiscretizado2Normalizado = pd.read_csv('C:/Users/John/Desktop/Mineracao/projetoBikes/bikesDiscretizado2.csv', sep=';');
bikesDiscretizado2Normalizado = normaliza(bikesDiscretizado2Normalizado, 'qtd', 8737)
bikesDiscretizado2Normalizado.to_csv('bikesDiscretizado2Normalizado.csv', sep=';', index=False)
```

```python
#Discretização da Coluna qtd
import pandas as pd
bikesConvertido1 = pd.read_csv('C:/Users/John/Desktop/Mineracao/projetoBikes/bikesConvertido.csv', sep=';');
bikesConvertido2 = pd.read_csv('C:/Users/John/Desktop/Mineracao/projetoBikes/bikesConvertido.csv', sep=';');
bikesDiscretizado1 = bikesConvertido1
bikesDiscretizado2 = bikesConvertido2

name = ['Very low','Low', 'Medium', 'High', 'Very High']
# print(pd.cut(bikesDiscretizado1['qtd'], 5, precision=0))
bins1 = [-1,398,796,1193,1591,1988]
# print(pd.qcut(bikesDiscretizado2['qtd'], 5, precision=0))
bins2 = [0,51,210,446,761,1988]

# bikesDiscretizado1.head()

bikesDiscretizado1['qtd'] = pd.cut(bikesDiscretizado1['qtd'], bins1, labels=name)
bikesDiscretizado1.to_csv('bikesDiscretizado1.csv', sep=';', index=False)
bikesDiscretizado2['qtd'] = pd.cut(bikesDiscretizado2['qtd'], bins2, labels=name)
bikesDiscretizado2.to_csv('bikesDiscretizado2.csv', sep=';', index=False)
```

```python
# Aplicação do PCA
import pandas as pd
from sklearn.decomposition import PCA
teste = pd.read_csv('C:/Users/John/Desktop/Mineracao/projetoBikes/bikesDiscretizado2Normalizado.csv', sep=';');
# Obtendo os nomes das colunas.
classe = teste['qtd']
cols = list(teste.columns)
# Removendo colunas que nao serao inclusas na reducao de dimensionalidade.
cols.remove('qtd')
# Instanciando um PCA. O parametro n_components indica a quantidade de dimensoes que a base original sera reduzida.
pca = PCA(n_components=9, whiten=True)
# Aplicando o pca na base original... O atributo 'values' retorna um numpy.array de duas dimensões (matriz) contendo apenas os valores numericos do DataFrame.
dados_pca = pca.fit_transform(teste[cols].values)
# O metodo fit_transform retorna outro numpy.array de dimensao numero_objetos x n_components.Apos isso, instancia-se um novo DataFrame contendo a base de dados original com dimensionalidade reduzida.
dados_pca = pd.DataFrame(dados_pca)
dados_pca.to_csv('bikesDiscretizado2Normalizado25.csv', sep=';', index=False)
```
